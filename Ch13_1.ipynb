###Phân tích so sánh

#### Tương quan

##### Giữa Barack Obama và Mitt Romney, ai nhận được nhiều quyên góp nhỏ (<$200) hơn?

```python
small = (
    fec.loc[(fec['contb_receipt_amt'] > 0) & (fec['contb_receipt_amt'] < 200) &
           (fec['cand_nm'].isin(['Obama, Barack', 'Romney, Mitt']))]
)

result = (small
          .groupby('cand_nm')['contb_receipt_amt']
          .agg(total_small_donations_usd='sum', number_of_small_donations='count')
          .sort_values('total_small_donations_usd', ascending=False))

print(result)
```

* Nhận xét: Barack Obama nhận được **nhiều quyên góp nhỏ (<$200)** hơn hẳn Mitt Romney, cả về **tổng số tiền** và **số lượt đóng góp**.
Điều này cho thấy chiến dịch của Obama có **mạng lưới ủng hộ rộng và phổ biến hơn** trong dân chúng.
Ngược lại, Romney chủ yếu dựa vào **các nhà tài trợ lớn** thay vì quyên góp nhỏ lẻ.


##### Nghề nghiệp nào đóng góp nhiều nhất cho mỗi bên?

```python
import pandas as pd

# Chỉ lấy 2 ứng viên và bỏ khoản âm (refund/phí)
two = fec.loc[
    (fec['cand_nm'].isin(['Obama, Barack', 'Romney, Mitt'])) &
    (fec['contb_receipt_amt'] > 0)
].copy()

# Chuẩn hoá occupation
two['contbr_occupation'] = (two['contbr_occupation']
                            .fillna('UNKNOWN')
                            .astype(str)
                            .str.strip()
                            .str.upper())

# Pivot: tổng tiền theo nghề nghiệp x ứng viên
pv = pd.pivot_table(
    two,
    index='contbr_occupation',
    columns='cand_nm',
    values='contb_receipt_amt',
    aggfunc='sum',
    fill_value=0
)

# Top 15 nghề đóng góp cho từng ứng viên
top_obama  = pv['Obama, Barack'].sort_values(ascending=False).head(15)
top_romney = pv['Romney, Mitt'].sort_values(ascending=False).head(15)

print("Top 15 nghề ủng hộ Obama:")
print(top_obama)
print("\nTop 15 nghề ủng hộ Romney:")
print(top_romney)

# Nghề đứng #1 cho mỗi bên
print("\nNghề #1 của Obama:", top_obama.index[0], "—", float(top_obama.iloc[0]))
print("Nghề #1 của Romney:", top_romney.index[0], "—", float(top_romney.iloc[0]))
```

* Cả Obama và Romney đều nhận được nhiều quyên góp nhất từ nhóm nghỉ hưu (RETIRED), phản ánh lượng cử tri lớn tuổi tích cực đóng góp.
Những nghề như luật sư, bác sĩ, giáo sư ủng hộ Obama mạnh hơn, thể hiện xu hướng nghiêng về giới trí thức và khu vực công.
Ngược lại, Romney nhận nhiều từ giới kinh doanh và tài chính như CEO, chủ tịch, nhà đầu tư, cho thấy sức ủng hộ từ khối doanh nghiệp và giới thượng lưu.


#### Giải thích

##### Có mối quan hệ nào giữa nghề nghiệp và định hướng chính trị không?

```python
import pandas as pd
import numpy as np

# Lọc 2 ứng viên, chỉ lấy khoản >0
twoc = fec.loc[
    (fec['cand_nm'].isin(['Obama, Barack','Romney, Mitt'])) &
    (fec['contb_receipt_amt'] > 0)
].copy()

# Chuẩn hoá occupation
twoc['occ'] = (twoc['contbr_occupation']
               .fillna('UNKNOWN').astype(str).str.upper().str.strip())

# Gom nhóm vài biến thể phổ biến
ALIAS = {
    'SOFTWARE ENGINEER':'ENGINEER', 'ENGINEER':'ENGINEER', 'ENG':'ENGINEER',
    'TEACHER':'TEACHER', 'PROFESSOR':'PROFESSOR', 'EDUCATOR':'TEACHER',
    'INVESTOR':'INVESTOR', 'INVESTMENT':'INVESTOR', 'VENTURE CAPITALIST':'INVESTOR',
    'ATTORNEY':'LAWYER', 'LAWYER':'LAWYER',
}
twoc['occ'] = twoc['occ'].replace(ALIAS)

# Loại các occupation “rác” nếu muốn nhìn nghề nghiệp thực
EXCLUDE = {
    'RETIRED','NOT EMPLOYED','NONE','SELF-EMPLOYED','SELF EMPLOYED',
    'UNEMPLOYED','HOMEMAKER','STUDENT','INFORMATION REQUESTED',
    'INFORMATION REQUESTED PER BEST EFFORTS','UNKNOWN'
}
occ_clean = twoc[~twoc['occ'].isin(EXCLUDE)].copy()

# Bảng tổng quyên góp theo nghề x ứng viên
occ_pv = pd.pivot_table(
    occ_clean, index='occ', columns='cand_nm',
    values='contb_receipt_amt', aggfunc='sum', fill_value=0
)

# Nghề “thiên” về ứng viên nào? (tỷ trọng phần trăm theo nghề)
share = occ_pv.div(occ_pv.sum(axis=1), axis=0)  # mỗi hàng chuẩn hoá 100%
lean_obama  = share.sort_values('Obama, Barack', ascending=False).head(20)
lean_romney = share.sort_values('Romney, Mitt', ascending=False).head(20)

print("Nghề nghiêng về Obama (tỷ trọng cao):")
print(lean_obama[['Obama, Barack','Romney, Mitt']].head(10))
print("\nNghề nghiêng về Romney (tỷ trọng cao):")
print(lean_romney[['Obama, Barack','Romney, Mitt']].head(10))
```

* Các nghề như ENGINEER, TEACHER, LAWYER, SCIENTIST có tỷ lệ ủng hộ Obama cao hơn, thể hiện xu hướng nghiêng về Đảng Dân chủ của giới trí thức và công nghệ.
Ngược lại, các nghề như INVESTOR, EXECUTIVE, FINANCE, BUSINESS OWNER lại đóng góp nhiều cho Romney, phản ánh sự ủng hộ mạnh từ giới kinh doanh và tài chính.
Như vậy, nghề nghiệp có mối quan hệ rõ ràng với định hướng chính trị, phản ánh sự phân hóa kinh tế – xã hội trong hành vi quyên góp.


##### Các ngành (công nghệ, giáo dục, tài chính) có xu hướng ủng hộ đảng nào?

```python
# Chuẩn hoá employer
twoc['emp'] = (twoc['contbr_employer']
               .fillna('UNKNOWN').astype(str).str.upper().str.strip())

# Tạo bộ luật phân loại đơn giản (tuỳ chỉnh thêm theo dữ liệu của bạn)
def map_industry(emp: str) -> str:
    if any(k in emp for k in ['GOOGLE','MICROSOFT','FACEBOOK','APPLE','AMAZON','IBM','ORACLE','ADOBE','INTEL']):
        return 'TECH'
    if any(k in emp for k in ['UNIVERSITY','COLLEGE','SCHOOL','INSTITUTE']):
        return 'EDUCATION'
    if any(k in emp for k in ['BANK','GOLDMAN','MORGAN','CITIGROUP','JPMORGAN','WELLS FARGO','FINANC','CAPITAL','VENTURE','PRIVATE EQUITY']):
        return 'FINANCE'
    if 'GOVERNMENT' in emp or 'CITY OF' in emp or 'STATE OF' in emp or 'US ' in emp:
        return 'GOVERNMENT'
    if any(k in emp for k in ['HOSPITAL','HEALTH','MED','CLINIC','PHARMA']):
        return 'HEALTHCARE'
    return 'OTHER'

twoc['industry'] = twoc['emp'].apply(map_industry)

ind_pv = pd.pivot_table(
    twoc, index='industry', columns='cand_nm',
    values='contb_receipt_amt', aggfunc='sum', fill_value=0
).sort_values(['Obama, Barack','Romney, Mitt'], ascending=False)

print(ind_pv)
```

* Ngành giáo dục (EDUCATION) và công nghệ (TECH) quyên góp nhiều hơn cho Barack Obama, thể hiện xu hướng ủng hộ Đảng Dân chủ của giới trí thức và công nghệ.
Ngược lại, ngành tài chính (FINANCE) và chăm sóc sức khỏe (HEALTHCARE) đóng góp mạnh cho Mitt Romney, phản ánh sự ủng hộ từ giới kinh doanh và tài chính đối với Đảng Cộng hòa.
Nhìn chung, các ngành tri thức và công nghệ nghiêng về Obama, còn ngành tài chính và doanh nghiệp nghiêng về Romney.


#### Mức quyên góp

```python
pd.cut(fec['contb_receipt_amt'], bins=[0, 100, 1000, 10000, 100000, 1000000])
```

Ngược lại, Mitt Romney nhận tỷ lệ cao hơn từ các khoản lớn trên $1.000, phản ánh sự ủng hộ mạnh từ giới tài trợ giàu có và doanh nghiệp.
Nhìn chung, Obama có nền tảng ủng hộ rộng rãi, còn Romney có ít người ủng hộ nhưng giá trị trung bình cao hơn.


##### Ứng cử viên nào có lượng người ủng hộ đông đảo hơn (nhiều người, ít tiền mỗi người)?

```python
# Xấp xỉ donor-unique bằng (tên, mã ZIP)
twoc['donor_id'] = (twoc['contbr_nm'].astype(str).str.upper().str.strip()
                    + '|' + twoc['contbr_zip'].astype(str).str[:5])

donor_stats = (twoc.groupby('cand_nm')
               .agg(unique_donors=('donor_id','nunique'),
                    total_amount=('contb_receipt_amt','sum')))
donor_stats['avg_per_donor'] = donor_stats['total_amount'] / donor_stats['unique_donors']
print(donor_stats.sort_values('unique_donors', ascending=False))
```

* Barack Obama có số lượng người ủng hộ nhiều gấp gần 3 lần so với Mitt Romney, cho thấy mạng lưới quyên góp rộng và phổ biến hơn.
 Tuy nhiên, mức quyên góp trung bình của Obama (≈$691/người) thấp hơn Romney (≈$1.319/người).
 Điều này cho thấy Obama thu hút nhiều người quyên góp nhỏ, còn Romney dựa vào ít người nhưng đóng góp lớn hơn.


#### Không gian địa lý

#####  Bang nào đóng góp nhiều nhất cho mỗi ứng cử viên?

```python
import pandas as pd

# Giả sử dữ liệu đã nằm trong DataFrame 'fec'

# Tính tổng tiền quyên góp theo ứng viên và bang
state_donations = (
    fec.groupby(['cand_nm', 'contbr_st'])['contb_receipt_amt']
    .sum()
    .reset_index()
)

# Lấy bang có quyên góp cao nhất cho từng ứng viên
top_state_each = (
    state_donations.sort_values(['cand_nm', 'contb_receipt_amt'], ascending=[True, False])
    .groupby('cand_nm')
    .head(1)   # lấy dòng đầu tiên (cao nhất) cho mỗi ứng viên
)

print(top_state_each)
```

* **California (CA)** là bang đóng góp nhiều nhất cho hầu hết các ứng viên, đặc biệt là **Barack Obama** và **Mitt Romney**.
**Texas (TX)** cũng nổi bật với mức quyên góp cao cho các ứng viên **Cộng hòa** như Perry và Santorum.
Điều này cho thấy **các bang đông dân, kinh tế mạnh** như CA và TX đóng vai trò chủ chốt trong huy động tài chính bầu cử.


##### Có sự khác biệt vùng miền (Bờ Đông / Bờ Tây) trong hành vi quyên góp không?

```python
east_states = {'ME','NH','MA','RI','CT','NY','NJ','PA','DE','MD','DC','VA','NC','SC','GA','FL'}
west_states = {'CA','OR','WA','AK','HI'}

twoc['coast'] = np.where(twoc['contbr_st'].isin(west_states), 'WEST COAST',
                         np.where(twoc['contbr_st'].isin(east_states), 'EAST COAST', 'OTHER'))

coast_pv = (twoc.groupby(['coast','cand_nm'])['contb_receipt_amt']
            .sum().unstack('cand_nm').fillna(0))
print(coast_pv)
```

* Kết quả cho thấy có sự khác biệt vùng miền rõ rệt trong hành vi quyên góp.
 Cả Bờ Đông (East Coast) và Bờ Tây (West Coast) đều quyên góp nhiều hơn cho Barack Obama, đặc biệt Bờ Đông dẫn đầu với hơn 59 triệu USD.
 Điều này phản ánh xu hướng truyền thống: Bờ Đông và Bờ Tây nghiêng về Đảng Dân chủ, trong khi các bang nội địa (Other) có xu hướng ủng hộ Mitt Romney nhiều hơn.
